# CDC PLACES Data Gather

CDC's filtered dataset for California has been uploaded to the volume in the bronze layer. The data dictionary can be found here - https://data.cdc.gov/500-Cities-Places/PLACES-Local-Data-for-Better-Health-Census-Tract-D/cwsq-ngmh/about_data

```python
# upload the cdc_places_ca_data to the bronze layer and read it back 
cdc_places_ca_spark_df = spark.read.csv("/Volumes/ca_healthcare_fac_bronze/default/ca_healthcare_volume_bronze/cdc_places_ca_data/PLACES__Local_Data_for_Better_Health,_Census_Tract_Data,_2025_release_20260411.csv", header=True)

print((cdc_places_ca_spark_df.count(), len(cdc_places_ca_spark_df.columns)))
```

```python
# Need to cast the columns to the correct data types
from pyspark.sql.functions import col, regexp_replace

cdc_places_ca_spark_df = cdc_places_ca_spark_df.withColumn("Data_Value", cdc_places_ca_spark_df.Data_Value.cast("double"))
cdc_places_ca_spark_df = cdc_places_ca_spark_df.withColumn("Low_Confidence_Limit", cdc_places_ca_spark_df.Low_Confidence_Limit.cast("double"))
cdc_places_ca_spark_df = cdc_places_ca_spark_df.withColumn("High_Confidence_Limit", cdc_places_ca_spark_df.High_Confidence_Limit.cast("double"))
cdc_places_ca_spark_df = cdc_places_ca_spark_df.withColumn("TotalPopulation", regexp_replace(col("TotalPopulation"), ",", "").cast("int"))
cdc_places_ca_spark_df = cdc_places_ca_spark_df.withColumn("TotalPop18plus", regexp_replace(col("TotalPop18plus"), ",", "").cast("int"))
```

```python
# Save the data to the bronze layer
cdc_places_ca_spark_df.write.mode("append").saveAsTable("ca_healthcare_fac_bronze.cdc_places_ca_data.cdc_places_ca_data")
```
